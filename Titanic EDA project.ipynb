```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df
```

```python
print(df.shape) # number of row & columns
print(df.columns) # column names
```

```python
df.head(10)
```

```python
print(df.info())
```

```python
df.describe()
```

```python
print(df.isnull())  # Missing values check
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True) # age's missing values handling
df.isnull().sum()
```

```python
df.drop('Cabin', axis=1, inplace=True) # Cabin column drop as there are so many null
df.isnull().sum()
```

```python
df.columns
```

```python
# Univariate Analysis
sns.countplot(x='Survived', data=df)
plt.title('Survived vs Not Survived')
plt.show()

sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
#Bivariate Analysis

sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Sex vs Survived')
plt.show()

sns.boxplot(x='Pclass', y='Age', hue='Survived', data=df)
plt.title('Pclass vs Age')
plt.show()
```

```python
# correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
